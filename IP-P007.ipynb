#### Métodos para processar várias colunas com informações sobre a temperatura 


```python
# Lista de listas para as colunas de temperatura
def calcular_media(coluna):
    return coluna.mean(skipna=True)
```

```python
def preencher_valores_nan_com_media(coluna):
    media = calcular_media(coluna)
    return coluna.fillna(media, inplace=True)
```

```python
def encontrar_menor_valor(coluna):
    return coluna.min(skipna=True)
```

```python
def encontrar_maior_valor(coluna):
    return coluna.max(skipna=True)
```

```python
def remover_valores_invalidos(coluna):
    return coluna.replace(-9999.0, None)
```

```python
def filtrar_por_mes(df, mes):
    return df.loc[(slice(None), mes, slice(None)), :]
```

#### Analisando as colunas por meio do loop for.

```python
def processar_colunas(df, colunas):
    for i in colunas:
        secao_nome, coluna_nome = i
        coluna = df[secao_nome][coluna_nome]
        coluna = remover_valores_invalidos(coluna)
        preencher_valores_nan_com_media(coluna)
        menor_valor = encontrar_menor_valor(coluna)
        maior_valor = encontrar_maior_valor(coluna)
        print(f'Coluna: {coluna_nome}')
        print(f'Média: {calcular_media(coluna)}')
        print(f'Menor valor: {menor_valor}')
        print(f'Maior valor: {maior_valor}')
        print(' ')
```

#### Exibindo temperaturas

```python
# Chamando a função para processar as colunas de temperatura
processar_colunas(df_concat, colunas_temperatura)
```

#### Analisando precipitações por estações do ano.

```python
def preencher_valores_nan_com_media(coluna):
    media = calcular_media(coluna)
    coluna_filled = coluna.fillna(media)
    return coluna_filled
```

```python
colunas_relacionadas = ['PRECIPITAÇÃO']
```

```python
estacoes = {
    'Verão': [1, 2, 3],
    'Outono': [4, 5, 6],
    'Inverno': [7, 8, 9],
    'Primavera': [10, 11, 12]
}
```

```python
def filtrar_por_estacao(df, estacao_indices):
    return df[df.index.get_level_values(1).isin(estacao_indices)]
```

```python
def calcular_medias_por_estacao(df, colunas, estacao_indices):
    medias_por_estacao = {}
    for estacao_nome, estacao_meses in estacao_indices.items():
        dados_estacao = filtrar_por_estacao(df, estacao_meses)
        medias_estacao = {}
        for coluna_nome in colunas:
            coluna = dados_estacao[coluna_nome].copy()
            remover_valores_invalidos(coluna)
            coluna_filled = preencher_valores_nan_com_media(coluna)
            media = calcular_media(coluna_filled)
            medias_estacao[coluna_nome] = media
        medias_por_estacao[estacao_nome] = medias_estacao
    return medias_por_estacao
```

```python
medias_por_estacao = calcular_medias_por_estacao(df_concat, colunas_relacionadas, estacoes)
```

#### Exibir estações do ano com suas, respectivas, médias de precipitações

```python
for estacao, medias in medias_por_estacao.items():
    print(f"Estação: {estacao}")
    for coluna, media in medias.items():
        media_float = media.mean()  # Calculando a média diretamente
        print(f"Média de {coluna}: {media_float:.5f} mm")
    print()
```
